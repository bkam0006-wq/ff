```python
import pandas as pd
import yfinance as yf
```

```python
# Load the Excel file
input_file = r"C:\Users\ShahreedzaAnissaMohd\repo\PyStockScreener\code_works\code_works\data\listed_co_on_bursa_131125.xlsx"  # <-- Updated to .xlsx
df_tickers = pd.read_excel(input_file)  # You can add sheet_name="Sheet1" if needed

#symbol, shortName, longName -- # USE FOR REFERENCE

# Format ticker from "Code"
def format_ticker(code):
    code = str(code).strip()
    if code.isdigit():
        return code.zfill(4) + ".KL"
    return code.upper()

df_tickers["Formatted_Ticker"] = df_tickers["Code"].apply(format_ticker)
tickers = df_tickers["Formatted_Ticker"].dropna().unique()

# Fetch data and append ticker manually
results = []
for ticker in tickers:
    # print(f"Trying: {ticker}")
    try:
        info = yf.Ticker(ticker).info
        if info and "longName" in info:
            info["ticker"] = ticker  # <-- Explicitly add the ticker
            results.append(info)
            # print(f"✓ {ticker}")
        else:
            print(f"✗ Not found: {ticker}")
    except Exception as e:
        print(f"✗ Error: {ticker} — {str(e)}")

# Save to CSV
if results:
    df_result = pd.DataFrame(results)
    df_result.to_excel(r"C:\Users\ShahreedzaAnissaMohd\repo\PyStockScreener\code_works\code_works\data\050126_co_listed_yfinance_data_.xlsx", index=False)
    print("\n✅ Data saved with 'ticker'")
else:
    print("\n⚠️ No valid data found.")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# --- Step 1: Load saved Yahoo Finance data ---
df = pd.read_csv(r"C:\Users\ShahreedzaAnissaMohd\code_works\data\co_listed_yfinance_data.csv")

# Optional: set index
if 'symbol' in df.columns: 
    df.set_index('symbol', inplace=True)

# Step 2: Keep numeric only
df_numeric = df.select_dtypes(include='number')

# Step 3: Drop columns with missing values (keep all rows)
df_clean = df_numeric.dropna(axis=1)

# Step 4: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean)

# Step 5: Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add cluster labels back to DataFrame
df_clean['Cluster'] = clusters

# Step 6: Visualize using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', s=100, edgecolors='k')
plt.title("K-Means Clustering of Malaysian Stocks")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")

# Annotate each point with ticker name
for i, ticker in enumerate(df_clean.index):
    plt.annotate(ticker, (X_pca[i, 0], X_pca[i, 1]), fontsize=8)

plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# ================================
# Step 1: Load Berkshire Hathaway Data
# ================================

berkshire_df = pd.read_csv(r"C:\Users\ShahreedzaAnissaMohd\code_works\data\bershire_hathaways_holdings_info.csv")

if 'Ticker' in berkshire_df.columns:
    berkshire_df.set_index('Ticker', inplace=True)

# Keep only numeric columns
berkshire_numeric = berkshire_df.select_dtypes(include='number')

# Drop columns with missing values
berkshire_clean = berkshire_numeric.dropna(axis=1)

# Save feature columns to reuse on Malaysia data
berkshire_features = berkshire_clean.columns.tolist()

# ================================
# Step 2: Standardize & Train KMeans on Berkshire
# ================================

scaler = StandardScaler()
X_berkshire_scaled = scaler.fit_transform(berkshire_clean)

kmeans = KMeans(n_clusters=3, random_state=42)
berkshire_clusters = kmeans.fit_predict(X_berkshire_scaled)

berkshire_clean['Cluster'] = berkshire_clusters

pca = PCA(n_components=2)
X_berkshire_pca = pca.fit_transform(X_berkshire_scaled)

# ================================
# Step 3: Load Malaysia Stock Data
# ================================

malaysia_df = pd.read_csv(r"C:\Users\ShahreedzaAnissaMohd\code_works\data\co_listed_yfinance_data.csv")

if 'Ticker' in malaysia_df.columns:
    malaysia_df.set_index('Ticker', inplace=True)
elif 'Stock' in malaysia_df.columns:
    malaysia_df.set_index('Stock', inplace=True)

# Keep only numeric columns
malaysia_numeric = malaysia_df.select_dtypes(include='number')

# Filter columns to match Berkshire's features (drop columns with NaNs)
malaysia_filtered = malaysia_numeric[berkshire_features].dropna(axis=1)

# Ensure final features match (fill missing columns with 0)
for col in berkshire_features:
    if col not in malaysia_filtered.columns:
        malaysia_filtered[col] = 0

# Reorder to match model input
malaysia_aligned = malaysia_filtered[berkshire_features]

# ================================
# Step 4: Transform & Classify Malaysia Stocks
# ================================

if malaysia_aligned.empty:
    raise ValueError("Malaysia data has no valid rows after aligning features. Check for excessive NaNs.")

X_malaysia_scaled = scaler.transform(malaysia_aligned)
malaysia_clusters = kmeans.predict(X_malaysia_scaled)
X_malaysia_pca = pca.transform(X_malaysia_scaled)

malaysia_aligned['Cluster'] = malaysia_clusters

# ================================
# Step 5: Visualize Malaysian Stocks in Berkshire PCA Space
# ================================

plt.figure(figsize=(10, 7))
plt.scatter(X_malaysia_pca[:, 0], X_malaysia_pca[:, 1], c=malaysia_clusters, cmap='plasma', s=100, edgecolors='k')
plt.title("Malaysian Stocks Classified by Berkshire-Trained KMeans")
plt.xlabel("PCA Component 1 (Berkshire space)")
plt.ylabel("PCA Component 2 (Berkshire space)")

for i, ticker in enumerate(malaysia_aligned.index):
    plt.annotate(ticker, (X_malaysia_pca[i, 0], X_malaysia_pca[i, 1]), fontsize=8)

plt.grid(True)
plt.tight_layout()
plt.show()

# ================================
# Step 6: Save Clustered Malaysian Stocks
# ================================

malaysia_aligned.to_csv(r"C:\Users\ShahreedzaAnissaMohd\code_works\data\result_malaysian_yfinance_data.csv")
```
